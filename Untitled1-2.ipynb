```python
import pandas as pd

df = pd.read_csv(list(uploaded.keys())[0])
df.head()
```

```python
df.columns = [
    "customer_id",
    "credit_lines",
    "loan_amount",
    "total_debt",
    "income",
    "years_employed",
    "fico_score",
    "default"
]

df.head()
```

```python
# Target (what we predict)
y = df["default"]

# Features (inputs)
X = df.drop(["default", "customer_id"], axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
def predict_expected_loss(data):

    input_df = pd.DataFrame([data])

    # Predict Probability of Default
    pd_value = model.predict_proba(input_df)[0][1]

    # Loan amount
    loan_amount = data["loan_amount"]

    # Expected Loss
    recovery_rate = 0.1
    expected_loss = pd_value * (1 - recovery_rate) * loan_amount

    return pd_value, expected_loss
```

```python
sample = {
    "credit_lines": 5,
    "loan_amount": 20000,
    "total_debt": 30000,
    "income": 50000,
    "years_employed": 5,
    "fico_score": 650
}

pd_val, loss = predict_expected_loss(sample)

print("Probability of Default:", pd_val)
print("Expected Loss:", loss)
```

```python
df_sorted = df.sort_values("fico_score").reset_index(drop=True)
```

```python
def create_initial_buckets(df, size=20):

    buckets = []

    for i in range(0, len(df), size):
        chunk = df.iloc[i:i+size]

        low = chunk["fico_score"].min()
        high = chunk["fico_score"].max()
        pd_val = chunk["default"].mean()

        buckets.append([low, high, pd_val])

    return buckets
```

```python
def merge_buckets(buckets, threshold=0.02):

    merged = []
    current = buckets[0]

    for b in buckets[1:]:

        # If PD difference small → merge
        if abs(current[2] - b[2]) < threshold:
            current[1] = b[1]  # extend range

            # update PD (average)
            current[2] = (current[2] + b[2]) / 2
        else:
            merged.append(current)
            current = b

    merged.append(current)

    return merged
```

```python
def get_rating_advanced(fico_score, buckets):

    for i, (low, high, pd_val) in enumerate(buckets):
        if low <= fico_score <= high:
            return i + 1, pd_val

    return len(buckets), None
```

```python
initial = create_initial_buckets(df_sorted, size=20)
final_buckets = merge_buckets(initial, threshold=0.02)

print("Final Buckets:")
for b in final_buckets:
    print(b)

print("\nTest:")
print(get_rating_advanced(720, final_buckets))
```
