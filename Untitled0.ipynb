```python
import pandas as pd

df = pd.read_csv(list(uploaded.keys())[0])
df.head()
```

```python
df.columns = ["date", "price"]
df["date"] = pd.to_datetime(df["date"])

df.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Create features
df["time_index"] = (df["date"] - df["date"].min()).dt.days
df["month"] = df["date"].dt.month

# Month dummies
month_dummies = pd.get_dummies(df["month"], prefix="month")

# Ensure all 12 months exist
for i in range(1, 13):
    col = f"month_{i}"
    if col not in month_dummies:
        month_dummies[col] = 0

month_dummies = month_dummies.sort_index(axis=1)

# Prepare data
X = pd.concat([df[["time_index"]], month_dummies], axis=1)
y = df["price"]

# Train model
model = LinearRegression()
model.fit(X, y)

# Prediction function
def predict_price(input_date):
    input_date = pd.to_datetime(input_date)

    time_index = (input_date - df["date"].min()).days
    month = input_date.month

    month_features = np.zeros(12)
    month_features[month - 1] = 1

    X_input = np.concatenate([[time_index], month_features]).reshape(1, -1)

    return float(model.predict(X_input)[0])

# Future prediction
future_dates = pd.date_range(df["date"].max(), periods=13, freq="M")[1:]
future_prices = [predict_price(d) for d in future_dates]

# Plot
plt.figure()
plt.plot(df["date"], df["price"], label="Historical")
plt.plot(future_dates, future_prices, "--", label="Forecast")
plt.legend()
plt.show()

# Test
print("Prediction:", predict_price("2025-06-30"))
```
