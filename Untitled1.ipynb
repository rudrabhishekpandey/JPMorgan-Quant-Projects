```python
import pandas as pd

df = pd.read_csv(list(uploaded.keys())[0])
df.head()
```

```python
df.columns = [
    "customer_id",
    "credit_lines",
    "loan_amount",
    "total_debt",
    "income",
    "years_employed",
    "fico_score",
    "default"
]

df.head()
```

```python
# Target (what we predict)
y = df["default"]

# Features (inputs)
X = df.drop(["default", "customer_id"], axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
def predict_expected_loss(data):

    input_df = pd.DataFrame([data])

    # Predict Probability of Default
    pd_value = model.predict_proba(input_df)[0][1]

    # Loan amount
    loan_amount = data["loan_amount"]

    # Expected Loss
    recovery_rate = 0.1
    expected_loss = pd_value * (1 - recovery_rate) * loan_amount

    return pd_value, expected_loss
```

```python
sample = {
    "credit_lines": 5,
    "loan_amount": 20000,
    "total_debt": 30000,
    "income": 50000,
    "years_employed": 5,
    "fico_score": 650
}

pd_val, loss = predict_expected_loss(sample)

print("Probability of Default:", pd_val)
print("Expected Loss:", loss)
```
